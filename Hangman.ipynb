```python
import random
import time
import requests

#Establish user parameters 
name = input("Enter your name: ")
difficulty = int(input("Choose length of word: "))
print("Good luck, " + name + "!")
print("\n")
time.sleep(2)
print("Instructions:")
time.sleep(1)
print("To start game, intialize the bottom block of code. Then run 'game()'")
```

```python
#core game function 
def game():
    global count
    global display
    global word
    global og_word
    global guessed
    global wlength
    global play
    words = "https://www.mit.edu/~ecprice/wordlist.10000"
    response = requests.get(words)
    wordbank = response.content.decode('utf-8').splitlines()
    #Initialize variables for game display 
    wlength = 0
    while wlength != difficulty:
        word = random.choice(wordbank) 
        wlength = len(word)

    count = 0
    display = '_' * wlength
    guessed= []
    play_game = ""
    word = word.lower()
    og_word = word
    hangman()
    
# A loop to re-execute the game when the first round ends:
def playAgain():
    global playA
    global difficulty
    playA = input("Play Again? \n y = yes, n = no \n")
    while playA not in ["y", "n","Y","N"]:
        playA = input("Invalid Input :(  \n Play Again?\n y = yes, n = no \n")
    playA = playA.lower() #Keep capital inputs as valid 
    if playA == "y":
        difficulty = int(input("Choose length of word to guess: "))
        game()
    elif playA == "n":
        print("Thanks for playing! :)")  
        
# Initializing all the conditions required for the game:
def hangman():
    global count
    global display
    global word
    global guessed
    global play_game
    limit = 6
    guess = input("Word to Guess: " + display + " Enter your guess: \n")
    guess = guess.strip().lower()
    if len(guess.strip()) == 0 or len(guess.strip()) >= 2 or guess.isalpha() == False: 
        print("Type only a single letter, please \n")
        hangman()    
    elif guess in word:
        guessed.append(guess)
        idx = word.find(guess)
        word = word[:idx] + "_" + word[idx + 1:]
        display = display[:idx] + guess + display[idx + 1:]
        print(display + "\n")
    elif guess in guessed:
        print("Letter already guessed.\n")
    else:
        guessed.append(guess)
        count += 1
        if count == 1:
            time.sleep(1)
            print("   _____ \n"
                  "  |     |\n"
                  "  |     |\n"
                  "  |     |\n"
                  "  |     O\n"
                  "  |      \n"
                  "  |      \n"
                  "__|__\n")
            print("Wrong -- " + str(limit - count) + " guesses remaining\n")
        elif count == 2:
            time.sleep(1)
            print("   _____ \n"
                  "  |     | \n"
                  "  |     |\n"
                  "  |     |\n"
                  "  |     O\n"
                  "  |     |\n"
                  "  |      \n"
                  "__|__\n")
            print("Wrong -- " + str(limit - count) + " guesses remaining\n")
        elif count == 3:
           time.sleep(1)
           print("   _____ \n"
                 "  |     | \n"
                 "  |     |\n"
                 "  |     | \n"
                 "  |     O \n"
                 "  |     |\ \n"
                 "  |      \n"
                 "__|__\n")
           print("Wrong -- " + str(limit - count) + " guesses remaining\n")
        elif count == 4:
            time.sleep(1)
            print("   _____ \n"
                  "  |     | \n"
                  "  |     |\n"
                  "  |     | \n"
                  "  |     O \n"
                  "  |    /|\ \n"
                  "  |      \n"
                  "__|__\n")
            print("Wrong -- " + str(limit - count) + " guesses remaining\n")
            
        elif count == 5:
            time.sleep(1)
            print("   _____ \n"
                  "  |     | \n"
                  "  |     |\n"
                  "  |     | \n"
                  "  |     O \n"
                  "  |    /|\ \n"
                  "  |    /  \n"
                  "__|__\n")
            print("Wrong -- " + " this is your final guess!\n")
        elif count == 6:
            time.sleep(1)
            print("   _____ \n"
                  "  |     | \n"
                  "  |     |\n"
                  "  |     | \n"
                  "  |     O \n"
                  "  |    /|\ \n"
                  "  |    / \ \n"
                  "__|__\n")
            print("You lose, " + name +  " :(\n")
            print("The word was:" + og_word)
            playAgain()
    if word == '_' * wlength:
        print("Congrats," +name+ "! You won!")
        playAgain()
    elif count != limit:
        hangman()
```

```python
#Run game here
game()
```
